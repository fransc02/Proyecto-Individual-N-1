Iniciamos con las Transformaciones

1. Generar campo id: Cada id se compondrá de la primera letra del nombre de la plataforma, seguido del show_id ya presente en los datasets (ejemplo para títulos de Amazon = as123)

```python
#Se crea el campo "id" agregando la primera letra de la plataforma
df_amazon['id'] = 'a' + df_amazon['show_id'] #
#Extraigo la  columna recien creada
first_column_a = df_amazon.pop('id') 
#Inserto el "id" como primera columna
df_amazon.insert(0,'id',first_column_a) 

#Se crea el campo "id" agregando la primera letra de la plataforma
df_hulu['id'] = 'h' + df_hulu['show_id'] 
#Extraigo la columna recien creada
first_column_h = df_hulu.pop('id') 
#Inserto el "id" como primera columna
df_hulu.insert(0,'id',first_column_h) 

#Se crea el campo "id" agregando la primera letra de la plataforma
df_netflix['id'] = 'n' + df_netflix['show_id'] 
#Extraigo columna recien creada
first_column_n = df_netflix.pop('id') 
#Inserto el "id" como primera columna
df_netflix.insert(0,'id',first_column_n) 

#Se crea el campo "id" agregando la primera letra de la plataforma
df_disney['id'] = 'd' + df_disney['show_id'] 
#Extraigo columna recien creada
first_column_d = df_disney.pop('id')
#Inserto el "id" como primera columna
df_disney.insert(0,'id',first_column_d) 
```

```python
#Se uninieron los csv's en uno solo
df_plataformas = pd.concat([df_amazon, df_netflix, df_disney,  df_hulu], axis=0) 
#Se elimina la columna "show_id"
df_plataformas.drop(["show_id"], axis = 1, inplace = True)  
```

2. Los valores nulos del campo rating deberán reemplazarse por el string “G” (corresponde al maturity rating: “general for all audiences”)


```python
#Se reemplaza los NaNs con la letra "G"
df_plataformas['rating'].fillna('G', inplace = True) 
```

3. De haber fechas, deberán tener el formato AAAA-mm-dd

```python
#Elimino la coma para que cumpla con el formato de "pd.to_datatime"
df_plataformas['date_added'] = df_plataformas['date_added'].replace({',':''}, regex=True) 
#Cambio la columna de "str" a "datetime64"
df_plataformas['date_added'] = pd.to_datetime(df_plataformas['date_added']) 
```

4. Los campos de texto deberán estar en minúsculas, sin excepciones

```python
#Se pasa a miniscula todos los strings
df_plataformas['type'] = df_plataformas['type'].str.lower() 
#Se pasa a miniscula todos los strings
df_plataformas['title'] = df_plataformas['title'].str.lower() 
#Se pasa a miniscula todos los strings
df_plataformas['director'] = df_plataformas['director'].str.lower() 
#Se pasa a miniscula todos los strings
df_plataformas['cast'] = df_plataformas['cast'].str.lower() 
#Se pasa a miniscula todos los strings
df_plataformas['country'] = df_plataformas['country'].str.lower() 
#Se pasa a miniscula todos los strings
df_plataformas['rating'] = df_plataformas['rating'].str.lower()
#Se pasa a miniscula todos los strings
df_plataformas['listed_in'] = df_plataformas['listed_in'].str.lower() 
#Se pasa a miniscula todos los strings
df_plataformas['duration'] = df_plataformas['duration'].str.lower() 
#Se pasa a miniscula todos los strings
df_plataformas['description'] = df_plataformas['description'].str.lower() 
```

5. El campo duration debe convertirse en dos campos: duration_int y duration_type. El primero será un integer y el segundo un string indicando la unidad de medición de duración: min (minutos) o season (temporadas)

```python
#Se crean los columna mediante la columna "duration" al separar sus strings por el " "(espacio) que hay entre ellos
df_plataformas[['duration_int', 'duration_type']] = df_plataformas['duration'].str.split(' ', n=-1, expand=True) 
#Se crea una variable para extraer la columna "duration_int"
columna_int = df_plataformas.pop('duration_int') 
#Insertamos la columna previamente extraida
df_plataformas.insert(10,'duration_int',columna_int) 
#Se crea una variable para extraer la columna "duration_type"
columna_type = df_plataformas.pop('duration_type') 
#Insertamos la columna previamente extraida
df_plataformas.insert(11,'duration_type',columna_type) 
#Reemplazmos todos los "seasons" por "season"
df_plataformas['duration_type'].replace("seasons","season", inplace = True) 
#Cambiamos la columna de "str" a "int64" 
df_plataformas['duration_int'] = df_plataformas['duration_int'].astype('Int64') 
#Eliminamos la columna "duration"
df_plataformas.drop('duration', axis = 1, inplace = True) 
```

Por ultimo convertimos nuestro dataframe a un cvs para nuestro archivo main.py

```python
df = pd.DataFrame(df_plataformas)
df.to_csv("plataformas.csv")
```
